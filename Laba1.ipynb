```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("D:/Download/spaceship-titanic/test.csv")

df.head()
```

```python
nan_matrix = df.isnull()
nan_matrix.sum()
```

```python
age_mean = df['Age'].mean()
df['Age'].fillna(age_mean, inplace = True)

home = df['HomePlanet'].mode()[0]
df['HomePlanet'].fillna(home, inplace = True)

Cryo_mode = df['CryoSleep'].mode()[0]
df['CryoSleep'].fillna(Cryo_mode, inplace = True)

dest = df['Destination'].mode()[0]
df['Destination'].fillna(dest, inplace = True)

vip_mode = df['VIP'].mode()[0]
df['VIP'].fillna(vip_mode, inplace = True)

room = df['RoomService'].mean()
df['RoomService'].fillna(room, inplace = True)

food_mean = df['FoodCourt'].mean()
df['FoodCourt'].fillna(food_mean, inplace = True)

shopping = df['ShoppingMall'].mean()
df['ShoppingMall'].fillna(shopping, inplace = True)

spa_mean = df['Spa'].mean()
df['Spa'].fillna(spa_mean, inplace = True)

vrdeck_mean = df['VRDeck'].mean()
df['VRDeck'].fillna(vrdeck_mean, inplace = True)

Name_mode = df['Name'].mode()[0]
df['Name'].fillna(Name_mode, inplace = True)

Cabin_mode = df['Cabin'].mode()[0]
df['Cabin'].fillna(Cabin_mode, inplace = True)
```

```python
nan_matrix = df.isnull()
nan_matrix.sum()
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])

scaler.fit(df[['RoomService']])
df['RoomService'] = scaler.transform(df[['RoomService']])

scaler.fit(df[['FoodCourt']])
df['FoodCourt'] = scaler.transform(df[['FoodCourt']])

scaler.fit(df[['ShoppingMall']])
df['ShoppingMall'] = scaler.transform(df[['ShoppingMall']])

scaler.fit(df[['Spa']])
df['Spa'] = scaler.transform(df[['Spa']])

scaler.fit(df[['VRDeck']])
df['VRDeck'] = scaler.transform(df[['VRDeck']])
```

```python
df.head()
```

```python
le = LabelEncoder()

df = pd.get_dummies(df, columns=['Destination'], drop_first=True)

df = pd.get_dummies(df, columns=['HomePlanet'], drop_first=True)

df = pd.get_dummies(df, columns=['CryoSleep'], drop_first=True)

df = pd.get_dummies(df, columns=['VIP'], drop_first=True)

df['Name'] = le.fit_transform(df['Name'])

df['Cabin'] = le.fit_transform(df['Cabin'])
```

```python
df.to_csv("processed_titanic.csv", index=False)
```
